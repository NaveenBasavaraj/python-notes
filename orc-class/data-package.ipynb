```python
import numpy as np
import matplotlib.pyplot as plt
house_price = np.array([1000, 2000, 3000, 4000, 5000])
sqft = np.array([500, 1000, 1500, 2000, 2500])
```

```python
plt.scatter(house_price, sqft, marker="+")
plt.xlabel("house price")
plt.ylabel("square feet")
plt.title("House Price Visualization")
```
